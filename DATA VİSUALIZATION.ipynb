```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.arange(20)
y=np.random.normal(10,1,20)#(n√º,sigma,20 tane)
z=np.random.normal(10,2,20)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,color="r")
ax.set_title("Line chart" )
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,label="Thing #1")
ax.plot(x,z,label="Thing #2")
ax.set_title("Line chart" )
ax.legend()

```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y,marker="o",c="#FF0000")
ax.set_xlabel("years",fontsize=20)
ax.set_ylabel("$$$money",fontsize=15)
fig.savefig("figure.png",dpi=300)
```

```python
fig2=plt.figure(figsize=(15,5))
a=fig2.add_subplot(131)
a.scatter(x,y)
a.set_title("Scatter plot")

b=fig2.add_subplot(132)
b.hist(y)
b.set_title("Histogram")

c=fig2.add_subplot(133)
c.boxplot([y,z])
c.set_title("Box plot")

fig2.suptitle("Vƒ∞SUALIZATION")
```
